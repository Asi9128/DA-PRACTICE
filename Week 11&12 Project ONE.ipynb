```python
#changing the amount($) column into string
join_data["Amount($)"] = join_data["Amount($)"].astype(str)
```

```python
# Replacing the dollar sign on each row with nothing usingn the lambda function

join_data["Amount($)"] = join_data["Amount($)"].apply(lambda x:x.replace("$",''))

#Replacing the ","sign with nothing using the lambda function
join_data["Amount($)"] = join_data["Amount($)"].apply(lambda x:x.replace(',',''))
```

```python
join_data["Amount($)"] = join_data["Amount($)"].apply(lambda x:x.replace("₹",''))


join_data["Amount($)"] = join_data["Amount($)"].apply(lambda x:x.replace("\t#REF",''))
```

```python
#apply 1 rupee to usd exchange
#rate = 0.014
# Define rup_index to find the indices where the "Amount" column contains the Rupee symbol (₹)
rup_index = join_data.index[join_data["Amount($)"].str.contains("₹", na=False)]

# Clean the "Amount" column: Remove the Rupee symbol and convert to numeric values
join_data.loc[rup_index, "Amount($)"] = join_data.loc[rup_index, "Amount($)"].str.replace("₹", "", regex=False).astype(float)

# Convert the amounts from Rupees to another currency (e.g., USD) using the conversion rate of 0.014
join_data.loc[rup_index, "Amount($)"] = join_data.loc[rup_index, "Amount($)"].values * 0.014

# Display the updated DataFrame
# print(join_data)
from IPython.display import display
display(join_data)
```

```python

#changing the amount($) column into string
# Step 2: Convert to float using pd.to_numeric with errors='coerce' to handle non-numeric values
join_data["Amount($)"] = pd.to_numeric(join_data["Amount($)"], errors='coerce')


join_data = join_data.dropna(subset=["Amount($)"])
# Verify the changes
print(join_data.dtypes)  # Output: 'Amount' column should now be float
print(join_data)
```
